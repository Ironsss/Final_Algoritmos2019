### Prueba de Gradient Descent 

Descenso por gradiente 

Es conocido como la medula espinal del Machine learning 

Son necesarias las paqueterias básicas de numpy y matplotlib para poder visualizar la información

funciones a utilizar en principio el problema 
a solucionar son encontrar cual de todas las regresiones hechas es la correcta para esto se utiiza una función conocida como mean squered error o error por promedio al cuadrado esta también es conocida como una función de costo 


Generar un descenso por gradiente consiste principalmente en ver como bajar por una pendiente mínima de acuerdo al concepto de derivada tenemos que será la tangente a una curva asociada por lo que debemos obtenr dicha tangente sin embargo está será obtenida con ayuda de un operador gradiente pudiendo tener un mínimo punto a punto

por lo tanto el método consiste en iterar la deruvada conforme avanzemos en la gráfica hasta encontrar un mínimo o un punto de inflexión 

teniendo el error del ajuste lineal de cada recta dada por:

$$ costo=  mse = \frac{1}{n} \sum_{i=1}^{n} (\triangle i)^{2}  $$

Debido a que desde la perspectiva del cálculo buscamos una distancia entre 2 puntos 

$$ mse = \frac{1}{n} \sum_{i=1}^{n} (y_{i} - y_{f(x)})^{2}  $$

susttituyendo $y_{f(x)}$ :
$$ mse = \frac{1}{n} \sum_{i=1}^{n} (y_{i} - (mx_{i} + b))^{2} $$

Derivando ambas ecuaciones parcialmente con respecto a las variables desconocidas m y b obtenidas por el método de regression:

$$ \frac{\partial (mse) }{\partial m} = \frac{2}{n} \sum_{i=1}^{n} -x_{i}(y_{i} - (mx_{i} + b)) $$
$$ \frac{\partial (mse) }{\partial b} = \frac{2}{n} \sum_{i=1}^{n} -(y_{i} - (mx_{i} + b)) $$

$$ \frac{\partial (mse) }{\partial m} = -\frac{2}{n} \sum_{i=1}^{n} x_{i}(y_{i} - (mx_{i} + b)) $$
$$ \frac{\partial (mse) }{\partial b} = -\frac{2}{n} \sum_{i=1}^{n} (y_{i} - (mx_{i} + b)) $$

considerando la siguiente igualdad debido a la construcción del método 

$learning rate = \alpha$
$$ m = m - \alpha*\frac{\partial (mse) }{\partial m} $$
$$ b = b - \alpha*\frac{\partial (mse) }{\partial b} $$


```python
def gradient_descent(x,y): #Construcción de la minimización por GradientDescent 
    inib = inim = 0  #Incisalizamos el contandor
    itera = 1000   #Número de iteraciones que desamos obtener 
    m = []        #creación de variables que deseamos obtener
    b = []
    n = len(x)    #número de acuerdo a la longitud de la lista tomada
    learning_rate = 0.000001  #nuestro alfa en este caso del modelo
    for i in range(itera):
        
        yf = inim*x + inib    #misma fórmula descrita previamente
        cost = (1/n)*sum((y - yf)**2)  # Costo definido 
        dm = -(2/n)*sum(x*(y-yf))  #nueva asignación de las derivadas de la pendiente e intersección
        db = -(2/n)*sum(y-yf)
        inim = inim - learning_rate*dm   #como obtenemos la nueva pendiente y ordenada
        inib = inib - learning_rate*db
        m.append(inim)  #guardamos las nuevas variables
        b.append(inib)
        plot(yf)     #graficamos
        #print("m{} b{} c:{} iteration{}".format(inim,inib,cost,i))
        print("c:{}".format(cost))
    #return m,b
    return inim,inib     #obtenemos los valores que nos interesan 

#al palicar el método pudimos reducir y minimizar la función de la regsión teniendo aquella que 
#tuviera el menor costo promedio probamos con un ajuste encillo para depués ajustar el modelo a los 
#ECG sin embargo observamos que no había mucha diferencia en el modelo del gradient descent construido al 
#modelo de sklearn para el ajuste lineal 
```

```python
X = 1.8*np.random.rand(100,1)
Y = 4 +X**2*np.random.randn(100,1)
```

```python
#Como en los casos anteriores se establece una distribución para comprender ocrrectamente el modelo y 
#y despues implementar el modelo a un ECG el proposito de hacer el ajuste polinomial para n grado
#es obtener algo caracterice los datos de cada pulso y no usar los datos en bruto 
#para esto proponemos tomar los coeficientes que acompañan al ajuste polinomial y hacer una DataFrame
#que contenga una sucesión de coeficientes por cada ECG

df = pd.DataFrame(X)
df2 = pd.DataFrame(Y)

x = df[0]
y = df2[0]

x1 = df[0]
y1 = df2[0]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
x_train2 = x_train.values.reshape([x_train.values.shape[0],1])
x_test2 = x_test.values.reshape([x_test.values.shape[0],1])

for i in range(1,10):
    PolN = PolynomialFeatures(degree = i)
    x_poln = PolN.fit_transform(x_train2) #valores en x
    xt_poln = PolN.fit_transform(x_test2) #valores en y?
    poly_model = LinearRegression()
    poly_model.fit(x_poln,y_train)
    
    x_pred = poly_model.predict(x_poln)
    xt_pred = poly_model.predict(xt_poln)
    x,y = zip(*sorted(zip(x_train2,x_pred)))
    xt,yt = zip(*sorted(zip(x_test,xt_pred)))
    plt.plot(x,y)
    
    
    
    #Lo que se realizó principalmente es una extensión del modelo de la regresión lineal para 
    #establecerlo se hizo uso de las funciones dadas por PolynomialFeatures de sklearn 
    #donde podemos extender el modelo a n grado deseadas ocupando valores de regresión variables 
```

### Pasando correctamente el modelo al ECG

a pesar de que para la red neuronal tenemos bases de datos ajustables para una red neuronal siendo ya hechos split previamente por los distribuidores nosottros debemos para cada caso generar nuestros propios conjuntos train para cada uno de los ECG puesto que si usaramos el DataFrame original Test o Train tendríamos un entrenamiento de la máquina sobre los mismos ECG siendo ambos ejes compartiendo datos similares y no obtendriamos una regresión de hecho por ahora no tengo certeza de que sea al usar el Train y Test dado por el dsitribuidor son embargo podemos hacerlo al usar un tiempo dado construido previamente usando el split usal <

```python
listas = list()
renueva =[]
for n in range(3):
    ppECG1 = electro1.iloc[n]
    tiempo_ECG = tipan[0] 
    #plot(tiempo_ECG,ppECG1)
    x_trainECG, x_etest, y_trainECG, y_etest = train_test_split(tiempo_ECG, ppECG1, test_size = 0.002)
    
    x_trainECG = ppECG1.values.reshape([ppECG1.values.shape[0],1])
    x_etest = x_etest.values.reshape([x_etest.values.shape[0],1])

    y_trainECG = y_trainECG.append(pd.Series(0.122005))

    for i in range(1,3):
        PolN = PolynomialFeatures(degree = i)
        x_poln = PolN.fit_transform(x_trainECG) #valores en x
        xt_poln = PolN.fit_transform(x_etest) #valores en y?
        poly_model = LinearRegression()
        poly_model.fit(x_poln,y_trainECG)
        betas = poly_model.coef_
        x_pred = poly_model.predict(x_poln)
        xt_pred = poly_model.predict(xt_poln)
        x,y = zip(*sorted(zip(x_trainECG,x_pred)))
        xt,yt = zip(*sorted(zip(x_etest,xt_pred)))
    plt.plot(x,y)
    renueva.append(betas)
    listas.append(renueva)
    renueva = []
    print("esto es renueva:")
    print(renueva)
    print("esto es betas:")
    print(listas)
    #plt.plot(tipan[0],ppECG1)
    
    
    #a pesar de que se puede aproximar de manera individual el ajuste polinomial para una disribución de 
    #puntos tenemos que la variabilidad del peso adecuado para cada uno de los ECG no se obtiene de manera
    #trivial y esta debe irse encontrando de manera experimental hasta enocntrar el peso que minimize 
    #adecuadamente el costo actualmente no contamos con la información necesaria de estdística que nos 
    #ayude a programas dicho peso por lo que al querer hacer una iteración sobre todos los ECG 
    #de la base de datos nos encontramos con errores en el ajuste.
    
    
    #pudimos solcuionar el problema optando por construir el ajsute polinomial más allá de lo 
    #que sklearn ofrece haciendo el modelo usual propuesto por el álgebra lineal.
    
```
